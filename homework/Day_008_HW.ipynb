```python
app_train.describe()
```

```python
plt.figure(figsize=(10,6))
y = app_train["CNT_CHILDREN"]
plt.hist(y,bins=y.max()-y.min(), log=True, width=0.8)
plt.xlabel("# of children", fontsize=20)
plt.xticks( np.arange(y.min(), y.max())+0.4,np.arange(y.min(), y.max()))
plt.show()
```

```python
plt.figure(figsize=(8,4))
mask = app_train["AMT_INCOME_TOTAL"] < app_train.describe().loc["75%"]["AMT_INCOME_TOTAL"]
plt.hist(app_train[mask]["AMT_INCOME_TOTAL"], bins=14, width=10000)
plt.xticks(np.arange(25000,200001,25000))
plt.xlabel("salary", fontsize=20)

x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.show()
```
